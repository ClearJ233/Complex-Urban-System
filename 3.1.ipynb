```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib
import matplotlib.pyplot as plt
```

```python
df_2020 = pd.read_csv('us-counties-2020.csv')
df_2021 = pd.read_csv('us-counties-2021.csv')
df_2022 = pd.read_csv('us-counties-2022.csv')
#df_2023 = pd.read_csv('us-counties-2023.csv')
#df_county = pd.read_csv('us-counties.csv')
```

```python
df_3Y = pd.concat([df_2020,df_2021,df_2022])
df_3Y
```

```python
len(df_3Y.fips.unique())
```

```python
data_3Y = df_3Y.copy()
data_3Y['city'] = data_3Y['state']+'_'+data_3Y['county']
len(data_3Y.city.unique())
```

```python
df_NYC = df_3Y[df_3Y['county']=='New York City']
df_NYC
```

```python
data_3Y
```

```python
data_3Y['date'] = pd.to_datetime(data_3Y['date'])
```

Descriptive Analysis

```python
data_3Y.sort_values(by=['city', 'date'], inplace=True)
```

```python
data_3Y['growth'] = data_3Y.groupby('city')['cases'].diff()
data_3Y['growth'] = data_3Y['growth'].fillna(0).astype(int)
data_3Y
```

```python
data_3Y.sort_values(by=['growth'], ascending=False, inplace=True)
```

```python
data_3Y.head(20)
```

```python
import random
```

```python
agg = list(data_3Y['city'].unique())
```

```python
data_agg = data_3Y.groupby(['city', pd.Grouper(key='date', freq='M')])['cases'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 8)) 
for city in random.sample(agg,5):
    city_data = data_agg[data_agg['city'] == city]
    plt.plot(city_data['date'], city_data['cases'], label=city)

plt.xlabel('Date')
plt.ylabel('Case Number')
plt.title('Time Series for 5 Random City')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data_sort = data_3Y.sort_values(by='date')
```

```python
plt.figure(figsize=(12, 8)) 
for city in random.sample(agg,5):
    city_data2 = data_sort[data_sort['city'] == city]
    plt.plot(city_data2['date'], city_data2['growth'], label=city)

plt.xlabel('Date')
plt.ylabel('Growth')
plt.title('Time Series for Each City')
plt.legend()
plt.grid(True)
plt.show()
```
