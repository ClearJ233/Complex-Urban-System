```python
merged_data = pd.merge(
    data1[['Recip_County', 'Recip_State','Date','Series_Complete_Yes','Series_Complete_Pop_Pct','Census2019']],
    df[['Recip_County', 'Recip_State','cases','deaths','Date']],
    on=['Recip_County', 'Recip_State','Date'],
    how='inner'
)
```

```python
merged_data['Infection_Rate']=merged_data['cases']/merged_data['Census2019']*100
merged_data['Deaths_Rate']=merged_data['deaths']/merged_data['Census2019']*100
merged_data['Series_Complete_Pop_Pct']=merged_data['Series_Complete_Yes']/merged_data['Census2019']*100
merged_data
```

Cluster & Classificaton

```python
X_mean = merged_data.Series_Complete_Pop_Pct.mean()
Y_mean = merged_data.Infection_Rate.mean()
sigma1 = (((merged_data.Series_Complete_Pop_Pct-X_mean)**2).sum()/len(merged_data.Series_Complete_Pop_Pct))**(1/2)
sigma2 = (((merged_data.Infection_Rate-Y_mean)**2).sum()/len(merged_data.Infection_Rate))**(1/2)
print(X_mean,Y_mean,sigma1,sigma2)
```

```python
X=merged_data[['Series_Complete_Pop_Pct','Infection_Rate']].values
```

```python
cent=np.asarray([[X_mean+sigma1,Y_mean-sigma2],[X_mean-sigma1,Y_mean+sigma2]])
plt.figure(figsize=(4,4))
y_t=np.sum((X-cent[0,:])**2,axis=1)**0.5<np.sum((X-cent[1,:])**2,axis=1)**0.5 #attach points to clustering based on centroid proximity
fig, ax = plt.subplots(figsize=(4,4))

scatter = ax.scatter(X[:,0],X[:,1],c=y_t,cmap=plt.cm.cool,alpha=0.5)
center = ax.scatter(cent[:,0],cent[:,1],c=[1,0],s=100,cmap=plt.cm.coolwarm)

legend1 = ax.legend(*scatter.legend_elements(),
                    loc="lower right", title="Cluster")
legend2 = ax.legend(*center.legend_elements(),
                    loc="upper right", title="Cluster\nCenter")
plt.gca().add_artist(legend1)
ax.set_title('Clusters with Centers')
```

```python
merged_data['labels1']=y_t+0
merged_data
```

```python
mdata=merged_data.iloc[:,:10]
mdata
```

```python
mdata = mdata.sort_values(by='Date')
```

```python
mdata['Diff_IR'] = mdata.groupby(['Recip_County','Recip_State'])['Infection_Rate'].diff().fillna(0)
```

```python
mdata['Diff_DT'] = mdata.groupby(['Recip_County','Recip_State'])['Date'].diff().dt.days.fillna(0)
mdata
```

```python
mdata['diff_by_day'] = mdata['Diff_IR']/mdata['Diff_DT']
```

```python
mdata = mdata.fillna(0)
```

```python
mdata.loc[mdata['diff_by_day'] >= 0.01, 'labels2'] = 0
mdata.loc[mdata['diff_by_day'] < 0.01, 'labels2'] = 1
```

```python
mdata
```

```python
X=mdata[['Series_Complete_Pop_Pct','Infection_Rate']]
Y=mdata.labels2
```

```python
fig, ax = plt.subplots(figsize=(4,4))

scatter = ax.scatter(X.iloc[:, 0], X.iloc[:, 1], s=50, c=Y, cmap=plt.cm.get_cmap('coolwarm', 2))

legend = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="Cluster")
ax.add_artist(legend)
ax.set_title('Clusters')
plt.show()
```

After the above clustering and analysis, we know that when evaluating the risk level of a city, we not only need to pay attention to the vaccine completion rate and infection rate of a certain city at the moment, but also the degree of change of the infection rate in the city is worthy of attention.

Do high-risk cities correspond to higher death tolls or rates?

If we believe that high-risk cities are also accompanied by higher daily death rates, which of the previous clustering labels should we choose, labels1 or labels2?

```python
merged_data = merged_data.sort_values(by='Date')
```

```python
merged_data['Diff_DR'] = merged_data.groupby(['Recip_County','Recip_State'])['Deaths_Rate'].diff().fillna(0)
merged_data['Diff_DT'] = merged_data.groupby(['Recip_County','Recip_State'])['Date'].diff().dt.days.fillna(0)
merged_data['deaths_diff'] = merged_data['Diff_DR']/merged_data['Diff_DT']
merged_data = merged_data.fillna(0)
```

```python
merged_data[merged_data['labels1']==0].deaths_diff.mean()
```

```python
merged_data[merged_data['labels1']==1].deaths_diff.mean()
```

```python
drdm = mdata['deaths_diff'].mean()
```

```python
LB1_0_accuracy = len(merged_data[(merged_data['deaths_diff']>=drdm)&(merged_data['labels1']==0)])/len(merged_data[merged_data['labels1']==0])
LB1_1_accuracy = len(merged_data[(merged_data['deaths_diff']<drdm)&(merged_data['labels1']==1)])/len(merged_data[merged_data['labels1']==1])
LB2_0_accuracy = len(mdata[(mdata['deaths_diff']>=drdm)&(mdata['labels2']==0)])/len(mdata[mdata['labels2']==0])
LB2_1_accuracy = len(mdata[(mdata['deaths_diff']<drdm)&(mdata['labels2']==1)])/len(mdata[mdata['labels2']==1])
```

```python
print('For the labels1, high risk accuracy is',LB1_0_accuracy,', low risk accuracy is',LB1_1_accuracy,'\nFor the labels2, high risk accuracy is',LB2_0_accuracy,'low risk accuracy is',LB2_1_accuracy)
```

```python
LB1_0_recall = len(merged_data[(merged_data['deaths_diff']>=drdm)&(merged_data['labels1']==0)])/len(merged_data[merged_data['deaths_diff']>=drdm])
LB1_1_recall = len(merged_data[(merged_data['deaths_diff']<drdm)&(merged_data['labels1']==1)])/len(merged_data[merged_data['deaths_diff']<drdm])
LB2_0_recall = len(mdata[(mdata['deaths_diff']>=drdm)&(mdata['labels2']==0)])/len(mdata[mdata['deaths_diff']>=drdm])
LB2_1_recall = len(mdata[(mdata['deaths_diff']<drdm)&(mdata['labels2']==1)])/len(mdata[mdata['deaths_diff']<drdm])
```

```python
print('For the labels1, high risk recall is',LB1_0_recall,', low risk recall is',LB1_1_recall,'\nFor the labels2, high risk recall is',LB2_0_recall,'low risk recall is',LB2_1_recall)
```
