```python
from scipy.integrate import solve_ivp
```

```python
def sir_model(t, y, beta, gamma, N):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return [dSdt, dIdt, dRdt]
```

```python
beta=0.2
gamma=0.1
N=10000
```

```python
I0 = 1
S0 = N - I0
R0 = 0
y0 = [S0, I0, R0]
t_s = (0, 200)
```

```python
sol = solve_ivp(sir_model, t_s, y0, args=(beta, gamma, N), t_eval=np.linspace(0, 200, 200))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(sol.t, sol.y[0], label='Susceptible')
plt.plot(sol.t, sol.y[1], label='Infected')
plt.plot(sol.t, sol.y[2], label='Recovered')
plt.xlabel('Time')
plt.ylabel('Number of Individuals')
plt.title('SIR Model Simulation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
len(sol.y[1])
```

```python
r0 = beta / gamma

re = S0 / N * beta / gamma

if re > 1:
    print("The disease is epidemic.")
else:
    print("The disease is not epidemic.")
```

```python
peak_I = np.max(sol.y[1])
print("Peak number of infections:", peak_I)
```

```python
steady_state_S = gamma*N/beta
print("Steady-state number of susceptible individuals:", steady_state_S)
```

```python
def equation(R, S0, N):
    Si = S0
    temp = S0 * np.exp(-R * (1 - Si / N))
    Si = temp
    return Si
```

```python
S0 = 0.9999 
N = 10000  
R = np.linspace(1, 5)
```

```python
S = [equation(r, S0, N) for r in R]
percentage = [1-i for i in S]
```

```python
plt.figure(figsize=(8, 6))
plt.plot(R, percentage, color='blue', linestyle='-')
plt.xlabel('R')
plt.ylabel('Percentage of Population Infected')
plt.title('Percentage of Population Infected vs. Reproductive Number')
plt.show()
```
